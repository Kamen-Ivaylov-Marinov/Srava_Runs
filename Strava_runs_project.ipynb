## Usefull Functions
This function stores all my runs from the Excel column **Distance** and, when called, filters the data for a specific year. 
```python
def distance(np_array, start_row, end_row):
```
This function follows the same logic as the one above, but now it extracts only the average speed from the column **Average Speed**.
```python

```
In this function, I extract all the dates from the Excel column **Date** and store them in a list with a format <u>*datetime.datetime*</u>
```python
def extract_date(np_array, start_row, end_row):
```

````python
def get_distance(np_array, start_row, end_row):
    np_array = np_array.iloc[start_row:end_row, 4:5]
    data_distance = np.array(np_array)
    return [float(i[0]) for i in data_distance]

def get_speed(np_array, start_row, end_row):
    np_array = np_array.iloc[start_row:end_row, 5:6]
    data_speed = np.array(np_array)
    return [float(f'{i[0]:.2f}') for i in data_speed]

def get_dates(np_array, start_row, end_row):
    np_array = np_array.iloc[start_row:end_row, :1]
    data_dates = np.array(np_array)
    date_list = [date_str for row in data_dates for date_str in row]
    dates = [datetime.strptime(d, '%b %d %Y %I:%M:%S %p') for d in date_list]
    return dates
````

# Data for 2023
Here is all the data that is needed for the year 2023

````python
distance_2023 = get_distance(df, 0, 77)
av_speed_2023 = get_speed(df, 0, 77)
dates_2023 = get_dates(df, 0, 77)
````

## Graphs
Here we will see the distance I ran on each day throughout 2023, with a total of 77 recorded runs.
The X-axis represents the **Date** of each run, while the Y-axis shows the **Distance** in kilometers.

````python
dates_2023, distance_2023 = zip(*sorted(zip(dates_2023, distance_2023)))
print(distance_2023)

plt.figure(figsize=(20, 7))
plt.plot(dates_2023, distance_2023, marker='D', markerfacecolor='r', markersize=4)
plt.xlabel('Dates')
plt.ylabel('Distance')
plt.title(f'Kilometers per day in 2023. Total runs: {len(distance_2023)}')
plt.tight_layout()
plt.grid(True)
plt.show()
````

The data reveals variability in training intensity, with some spikes showing long runs, particularly noticeable in May.
Toward the end of the year, the running distances appear more consistent but shorter overall.

### Average speed for 5km
This chart shows my running speed for distances between 4 and 6 kilometers during 2023, with a total of 17 runs.
The X-axis represents the **Date** of each run, and the Y-axis shows the **Average Speed** in kilometers per minute (1km/mm).

````python
dates_2023, distance_2023 = zip(*sorted(zip(dates_2023, distance_2023)))
dates_2023, av_speed_2023 = zip(*sorted(zip(dates_2023, av_speed_2023)))
distanc–µ_speed = list(zip(distance_2023, av_speed_2023))
combined_data = {}
for key, value in zip(dates_2023, distanc–µ_speed):
    combined_data[key] = value

dates_km5 =[]
speed_km5 = []
for key, value in combined_data.items():
    distance = value[0]
    speed = value[1]
    if 4 < distance <6:
        dates_km5.append(key)
        speed_km5.append(speed)

plt.figure(figsize=(20, 7))
plt.plot(dates_km5 , speed_km5 , marker = '^', markerfacecolor = 'r', markersize = 8)
plt.xlabel('Dates')
plt.ylabel('Average Speed (min/km)')
plt.title(f'Running speed for 4‚Äì6 km distances in year 2023. Total runs: {len(speed_km5)}')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
````

The data reveals some variability in pace, with several peaks in May and a noticeable drop toward the end of the year. Overall trend is downwards, which means that I am getting faster.
